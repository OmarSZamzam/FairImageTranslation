```python
import numpy as np
import pandas as pd
import csv
import os
```

```python
with open('/scratch1/zamzam/HCP_1200.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # skip the header row
    HCP_info = [[row[0], row[4], row[3]] for row in reader]
HCP_info = np.array(HCP_info);

df = pd.read_csv('/scratch1/zamzam/CAMCAN.csv', header=None, skiprows=1)
CamCan_info = df.iloc[:, :3].apply(lambda x: x.str[2:] if x.name == 0 else x.str[0] if x.name == 2 else x).to_numpy()
```

```python
data = {
    'img_path': [],
    'ID': [],
    'age': [],
    'Dataset': [],
    'sex': []
}
df = pd.DataFrame(data)


trainHCP = os.listdir('/scratch1/zamzam/HCP_nt_train')
for i in range(len(trainHCP)):
    trainHCP[i] = '/scratch1/zamzam/HCP_1200/' + trainHCP[i][:-3] + '/T1w/T1w_acpc_dc_restore_brain.nii.gz'
trainCamCan = os.listdir('/scratch1/akrami/CamCan_all_versions/CAMCAN_nt_train')
for i in range(len(trainCamCan)):
    trainCamCan[i] = '/scratch1/akrami/Latest_Data/camcan_preproc/' + trainCamCan[i][:-3] + '/T1mni.nii.gz'
    
trainHCP = np.array(trainHCP).reshape(-1, 1)

trainCamCan = np.array(trainCamCan).reshape(-1, 1)

for i in range(len(trainHCP)):
    for j in range(len(HCP_info)):
        if trainHCP[i,0][26:32] == HCP_info[j,0]:
            new_row = { 'img_path': [trainHCP[i,0], trainHCP[i,0].replace('T1w_', 'T2w_')],
                        'ID': HCP_info[j,0],
                        'age': float(HCP_info[j,1][:2]),
                        'Dataset': 'HCP',
                        'sex': HCP_info[j,2]
                      }
            df = df.append(new_row, ignore_index=True)
            break
for i in range(len(trainCamCan)):
    for j in range(len(CamCan_info)):
        if trainCamCan[i,0][50:56] == CamCan_info[j,0]:
            new_row = { 'img_path': [trainCamCan[i,0], trainCamCan[i,0].replace('T1mni','T2mni')],
                        'ID': CamCan_info[j,0],
                        'age': CamCan_info[j,1],
                        'Dataset': 'CAMCAN',
                        'sex': CamCan_info[j,2]
                      }
            df = df.append(new_row, ignore_index=True)
            break

df.to_csv('/scratch1/zamzam/train.csv', index=False)
```
